```python
# 데이터프레임 전처리하는 함수
def preprocess_dataframe(df):
    df = df.iloc[1:, 1:]
    df.columns = ['감정단어', '장면', '움직임관련', '소리관련', '메타포(비유)', '시각적특징', '대명사', '사람관계', '기타', '헷갈리는거']
    return df
```

```python
one_music = preprocess_dataframe(one_music_pre)
one_music.to_excel('two_music.xlsx')
```

```python
one_dance = preprocess_dataframe(one_dance_pre)
one_dance.to_excel('two_dance.xlsx')
```

```python
one_face = preprocess_dataframe(one_face_pre)
one_face.to_excel('two_face.xlsx')
```

```python
one_movie = preprocess_dataframe(one_movie_pre)
one_movie.to_excel('two_movie.xlsx')
```

```python
one_musicvideo = preprocess_dataframe(one_musicvideo_pre)
one_musicvideo.to_excel('two_musicvideo.xlsx')
```

```python
one_visualize = preprocess_dataframe(one_visualize_pre)
one_visualize.to_excel('two_visualize.xlsx')
```

```python
# 시각자료 4종 visuo 로 합치기
one_visuo = pd.concat([one_face, one_movie, one_musicvideo, one_visualize])
one_visuo.to_excel('two_visuo.xlsx')
one_visuo
```
