```python
# 콘텐츠 모달리티가 다 섞여있으므로, 딕셔너리에서 콘텐츠 모달리티를 찾아 모달리티별로 정리해 엑셀로 만들기

# 빈 데이터프레임에 따로 저장하기
empty_music = pd.DataFrame(index=range(0, 100), columns=['감정단어', '장면', '움직임관련', '소리관련', '메타포(비유)', '시각적특징', '대명사', '사람관계', '기타', '헷갈리는거'])
empty_dance = pd.DataFrame(index=range(0, 100), columns=['감정단어', '장면', '움직임관련', '소리관련', '메타포(비유)', '시각적특징', '대명사', '사람관계', '기타', '헷갈리는거'])
empty_face = pd.DataFrame(index=range(0, 100), columns=['감정단어', '장면', '움직임관련', '소리관련', '메타포(비유)', '시각적특징', '대명사', '사람관계', '기타', '헷갈리는거'])
empty_movie = pd.DataFrame(index=range(0, 100), columns=['감정단어', '장면', '움직임관련', '소리관련', '메타포(비유)', '시각적특징', '대명사', '사람관계', '기타', '헷갈리는거'])
empty_musicvideo = pd.DataFrame(index=range(0, 100), columns=['감정단어', '장면', '움직임관련', '소리관련', '메타포(비유)', '시각적특징', '대명사', '사람관계', '기타', '헷갈리는거'])
empty_visualize = pd.DataFrame(index=range(0, 100), columns=['감정단어', '장면', '움직임관련', '소리관련', '메타포(비유)', '시각적특징', '대명사', '사람관계', '기타', '헷갈리는거'])
```

```python
import itertools
```

```python
def find_word_and_add(dictionary, df, empty): # dictionary ---> empty 콘텐츠 모달리티를 지정하기
    
    # 원래 raw 딕셔너리를 가져와서 1차원 리스트로 만들어줌
    dictionary = dictionary.values.tolist()
    dictionary = list(itertools.chain.from_iterable(dictionary))
    dictionary = set(dictionary)
    
    for i in range(len(df)):
        if df.iloc[i, 0] in dictionary:
            empty.loc[i, '감정단어'] = df.iloc[i, 0]
            
    for i in range(len(df)):
        if df.iloc[i, 1] in dictionary:
            empty.loc[i, '장면'] = df.iloc[i, 1]
            
    for i in range(len(df)): 
        if df.iloc[i, 2] in dictionary:
            empty.loc[i, '움직임관련'] = df.iloc[i, 2]
            
            
    for i in range(len(df)):
        if df.iloc[i, 3] in dictionary:
            empty.loc[i, '소리관련'] = df.iloc[i, 3]
            
    for i in range(len(df)):       
        if df.iloc[i, 4] in dictionary:
            empty.loc[i, '메타포(비유)'] = df.iloc[i, 4]
            
    for i in range(len(df)):
        if df.iloc[i, 5] in dictionary:
            empty.loc[i, '시각적특징'] = df.iloc[i, 5]
            
    # 대체 왜 이거 두개를 for 문 나눠야 되는지 모르겠지만... 암튼 됨
    
    for i in range(len(df)):
        if df.iloc[i, 6] in dictionary:
            empty.loc[i, '대명사'] = df.iloc[i, 6]
            
    for i in range(len(df)):
        if df.iloc[i, 7] in dictionary:
            empty.loc[i, '사람관계'] = df.iloc[i, 7]
            
    for i in range(len(df)): 
        if df.iloc[i, 8] in dictionary:
            empty.loc[i, '기타'] = df.iloc[i, 8]
            
    for i in range(len(df)):
        if df.iloc[i, 9] in dictionary:
            empty.loc[i, '헷갈리는거'] = df.iloc[i, 9]
            
    return empty
```

```python
# 음악 
two_music = find_word_and_add(music_dict, two_pre_change, empty_music)
two_music.to_excel('two_music.xlsx')
```

```python
# 댄스
two_dance = find_word_and_add(dance_dict, two_pre_change, empty_dance)
two_dance.to_excel('two_dance.xlsx')
```

```python
# 표정
two_face = find_word_and_add(face_dict, two_pre_change, empty_face)
# two_face.to_excel('two_face.xlsx')
```

```python
# 영화
two_movie = find_word_and_add(movie_dict, two_pre_change, empty_movie)
# two_movie.to_excel('two_movie.xlsx')
```

```python
# 뮤직비디오
two_musicvideo = find_word_and_add(musicvideo_dict, two_pre_change, empty_musicvideo)
# two_musicvideo.to_excel('two_musicvideo.xlsx')
```

```python
# 시각화
two_visualize = find_word_and_add(visualize_dict, two_pre_change, empty_visualize)
# two_visualize.to_excel('two_visualize.xlsx')
```

```python
two_visuo = pd.concat([two_face, two_movie, two_musicvideo, two_visualize])
two_visuo.to_excel('two_visuo.xlsx')
```
