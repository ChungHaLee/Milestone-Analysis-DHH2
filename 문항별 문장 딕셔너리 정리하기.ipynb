## 감정단어+차원 vs 텍스트로 나누기

```python
# 텍스트 리스트만 뽑아내는 함수
def extract_text(df):
    lst = df['answer'].tolist()
    text_lst = []
    
    for i in range(len(lst)):
        try:
            if "select" not in lst[i]:
                lst[i] = lst[i].replace('[', '').replace(']', '')
                text_lst.append(lst[i])
        except:
            text_lst.append(lst[i])
        
    return text_lst
```

## 문항별로 모아서 문장 출력하기

```python
def make_dataframe(lst):
    num_1_lst = []
    num_2_lst = []
    num_3_lst = []
    num_4_lst = []
    num_5_lst = []
    num_6_lst = []
    
    for i in range(len(lst)):
        try:
            if len(lst[i].split('", "')) == 2:
                num_1 = lst[i].split('", "')[0]
                num_2 = lst[i].split('", "')[1]
                
            else:
                num_3 = lst[i].split('", "')[0]
                num_4 = lst[i].split('", "')[1]
                num_5 = lst[i].split('", "')[2]
                num_6 = lst[i].split('", "')[3]
                
                num_1_lst.append(num_1)
                num_2_lst.append(num_2)
                num_3_lst.append(num_3)
                num_4_lst.append(num_4)
                num_5_lst.append(num_5)
                num_6_lst.append(num_6)
                
        except:
            pass
            
            
    df = pd.DataFrame([num_1_lst, num_2_lst, num_3_lst, num_4_lst, num_5_lst, num_6_lst], index=['1번', '2번', '3번', '4번', '5번', '6번'])
    
    return df.transpose()
```

```python
make_dataframe(extract_text(music_df)).to_excel('music_by_numbers.xlsx')
```

```python
make_dataframe(extract_text(dance_df)).to_excel('dance_by_numbers.xlsx')
```

```python
make_dataframe(extract_text(visual_df)).to_excel('visual_by_numbers.xlsx')
```
