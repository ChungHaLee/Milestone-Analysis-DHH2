## 청인

```python
users_hearing = users[users['display_type']=='cheongin']
users_hearing = users_hearing[users_hearing['name'] != 'ㅇ']
users_hearing = users_hearing[users_hearing['name'] != 'ㄹ']
users_hearing = users_hearing[users_hearing['name'] != 's']
users_hearing = users_hearing[users_hearing['name'] != '최유진']
users_hearing
```

```python
# 청인 고유 아이디 리스트
users_hearing_id = users_hearing['user_id'].tolist()
print(users_hearing_id)
```

```python
# 다른 고유 아이디여도 두번 접속한 사람은 같은 사람이기 때문에
# 이름이 중복해서 등장하는 사람이 누구이며, 몇번 재접속하는지 인지하는 함수를 만들어줌

def find_duplicate_names(users):
    name_lst = []
    duplicate_count = {}
    
    for i in range(len(users)):
        name_lst.append(users.iloc[i])
        
    for j in name_lst:
        try: 
            duplicate_count[j] += 1
        except:
            duplicate_count[j] = 1
            
    return duplicate_count
```

```python
def find_name_lst(users):
    name_lst = []
    
    for i in range(len(users)):
        name_lst.append(users['name'].iloc[i])

    name_lst = set(name_lst) # 여기를 고치자!~
    name_lst = list(name_lst)
    
    return name_lst
```

```python
def name_with_ids(users):
    id_lst = []
    name_dict = {}
    name_lst = find_name_lst(users)
    
    for name in name_lst:
        ids_lst = users['user_id'][users['name'] == name]
        ids_lst = list(ids_lst)
        name_dict[name] = ids_lst
    
    return name_dict
```

```python
# 청인 참여자 재접속 횟수
find_duplicate_hearings = find_duplicate_names(users_hearing['name'])
print(find_duplicate_hearings)
```

```python
# 청인 참여자 이름과 그 사람의 고유 아이디 알기
find_hearings_name_with_ids = name_with_ids(users_hearing)
find_hearings_name_with_ids
```

### 데이터프레임 합쳐서 참여자명.csv 로 real_results 에 저장하기

```python
def find_name_and_concat_df(name, name_ids_dict):
    for key, value in name_ids_dict.items():
        if key == name:
            if len(value) == 1: # 1번 접속
                try:
                    df1 = pd.read_csv('./results/%s/answers1.csv'%value[0])
                    
                except:
                    df1 = pd.read_csv('./results/%s/answers2.csv'%value[0])
                    
                return df1
            
            elif len(value) == 2: # 2번 접속
                try:
                    df1 = pd.read_csv('./results/%s/answers1.csv'%value[0])
                    df2 = pd.read_csv('./results/%s/answers1.csv'%value[1])
                    
                except:
                    df1 = pd.read_csv('./results/%s/answers2.csv'%value[0])
                    df2 = pd.read_csv('./results/%s/answers2.csv'%value[1])
                    
                concat_df = pd.concat([df1, df2], ignore_index=True)
                return concat_df
            
            
            # 여기 함수 고치기
            elif len(value) == 3: # 3번 접속
                try:
                    df1 = pd.read_csv('./results/%s/answers1.csv'%value[0])
                    df2 = pd.read_csv('./results/%s/answers1.csv'%value[1])
                    df3 = pd.read_csv('./results/%s/answers1.csv'%value[2])
                    
                except:
                    df1 = pd.read_csv('./results/%s/answers2.csv'%value[0])
                    df2 = pd.read_csv('./results/%s/answers2.csv'%value[1])
                    df3 = pd.read_csv('./results/%s/answers2.csv'%value[2])
                
                concat_df = pd.concat([df1, df2, df3], ignore_index=True)
                
                return concat_df
            
            elif len(value) == 4: # 4번 접속
                try:
                    df1 = pd.read_csv('./results/%s/answers1.csv'%value[0])
                    df2 = pd.read_csv('./results/%s/answers1.csv'%value[1])
                    df3 = pd.read_csv('./results/%s/answers1.csv'%value[2])
                    df4 = pd.read_csv('./results/%s/answers1.csv'%value[3])
                    
                except:
                    df1 = pd.read_csv('./results/%s/answers2.csv'%value[0])
                    df2 = pd.read_csv('./results/%s/answers2.csv'%value[1])
                    df3 = pd.read_csv('./results/%s/answers2.csv'%value[2])
                    df4 = pd.read_csv('./results/%s/answers2.csv'%value[3])
                    
                concat_df = pd.concat([df1, df2, df3, df4], ignore_index=True)
                
                return concat_df
```

```python
def concat_every_names(users, find_name_with_ids):
    name_lst = find_name_lst(users)
    for p_name in name_lst:
        concated_df = find_name_and_concat_df(p_name, find_name_with_ids) # 참여자 이름/ 참여자 고유 ID 딕셔너리
    
    # 해당 데이터프레임을 .csv 로 real_results 폴더에 저장.
    # 파일 이름은 참여자 이름으로 하기
        concated_df.to_csv('./real_results/real_hear/real_hear_%s.csv'%p_name)
    
    return
```

```python
# 청인- 참여자 이름별로 합치기
concat_every_names(users_hearing, find_hearings_name_with_ids)
```

```python
# 참여자별 csv 를 한글로 바꾸기
def change_csv_into_korean(users):
    name_lst = find_name_lst(users)
    for p_name in name_lst:
        
        df = pd.read_csv('./real_results/real_hear/real_hear_%s.csv'%p_name)
        
        for i in range(len(df)):
            try:
                df['modality'][i] = df['modality'][i].encode('utf-8').decode('unicode_escape')
                df['answer'][i] = df['answer'][i].encode('utf-8').decode('unicode_escape')
            except:
                pass
    
        df.to_csv('./real_results/real_hear/real_hear_%s.csv'%p_name)
        
    return 
```

```python
# 모든 csv 파일을 하나에 합치기
def concat_everything():
    all_file_list = glob.glob(os.path.join('./real_results/real_hear', 'real_hear_*'))
    all_data = []
    for file in all_file_list:
        df = pd.read_csv(file)
        all_data.append(df)
        
    data_combine = pd.concat(all_data, axis=0, ignore_index=True)
    data_combine.to_csv('./real_results/concat_everything_hear.csv', index=False)
    return 
```
